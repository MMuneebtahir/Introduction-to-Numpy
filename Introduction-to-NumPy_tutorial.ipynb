#### Dot product example (nut butter sales)

```python
np.random.seed(0)
sales_amounts = np.random.randint (20,size=(5,3))
sales_amounts
```

```python
# create a weekly sales DataFrame
weekly_sales =pd.DataFrame(sales_amounts,index=['Mon','Tues','Wed', 'Thrus','Fri'],
                           columns=["Almod-butter","penut-butter", "cashew-butter"])
weekly_sales
```

```python
prices = np.array([10,8,12])
prices
```

```python
# prices datframe
butter_prices = pd.DataFrame(prices.reshape(1,3),
                            index=["prices"],
                             columns=["Almod-butter","penut-butter", "cashew-butter"])
butter_prices
```

```python
prices.shape
```

```python
sales_amounts.shape
```

```python
# shapes are not allign we have to transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
butter_prices.shape , weekly_sales.shape
```

```python
butter_prices.shape , weekly_sales.T.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
#  adding daily sales at the end of table
weekly_sales.shape, daily_sales.T.shape
```

```python
# create new column in pandas
weekly_sales["Total ($)"]= daily_sales.T
weekly_sales
```
